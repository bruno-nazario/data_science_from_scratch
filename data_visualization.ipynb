```python
import matplotlib.pyplot as plt

# Grafico simples introducao 
years = [2000, 2001, 2002, 2003, 2004, 2005]
gpd = [ 100, 200, 220, 333, 577, 784]

plt.plot(years, gpd, color = 'pink', linestyle = 'solid')
plt.title('Nominal GPD')
plt.ylabel('Billions of $')
plt.show()
```
